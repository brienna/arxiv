```python
metadata_df = pd.read_csv('arxiv_metadata_astroph.csv', 
                       dtype={'filename': str,
                              'filename_parsed': str,
                              'identifier': str,
                              'updated': str,
                              'doi': str}, 
                       parse_dates=['date_retrieved'])
```

Ensure that none of the parsed filenames are the same, somehow:

```python
unique, counts = np.unique(metadata_df['filename_parsed'], return_counts=True)
for i in counts:
    if i > 1:
        print(True)
```

Get ids of XML files downloaded & converted:

```python
arxiv_ids = []

for file in os.listdir('/Volumes/BRIENNAKH 4/xml'):
    if file.endswith('.xml'):
        this_id = os.path.splitext(os.path.basename(file))[0]
        arxiv_ids.append(this_id)
        
print('Num of XML files downloaded & converted: ' + str(len(arxiv_ids)))
```

Compare those ids with the metadata to find out how many papers failed to download & convert:

```python
# SO SLOW, DON'T RUN
failed_papers = [x for x in set(metadata_df['filename_parsed']) if x not in set(arxiv_ids)]
```

```python
xml_ids_formatted = pd.Index(np.array(arxiv_ids))
metadata_ids_formatted = pd.Index(np.array(metadata_df['filename_parsed']))
print('Num of XML ids: ' + str(len(xml_ids_formatted)))
print('Num of metadata ids: ' + str(len(metadata_ids_formatted)))
```

```python
%%time
comp1 = [_ for _ in xml_ids_formatted if _ not in metadata_ids_formatted]
print('Num of XML papers that don\'t have their ids in the metadata: ' + str(len(comp1)))
```
